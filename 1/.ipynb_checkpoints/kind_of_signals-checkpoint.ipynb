## 1. Аналоговый сигнал

```python
from sympy import symbols
from sympy import pi
from sympy import sin
from sympy.plotting import plot
```

```python
A = 2
fz = pi / 4
f = 1
t = symbols('t')
sig = A * sin(2 * pi * t + fz)
```

```python
p = plot(sig, (t, -4, 4))
```

## 2. Дискретный и квантованный сигнал

```python
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
A = 2
fz = np.pi / 4
f = 1
n = 160
t2 = np.linspace(-4, 4, n, endpoint=True)
wave = A * np.sin(2 * np.pi * t2 + fz)
```

```python
# построение графика
fig = plt.figure(figsize=(8, 4), dpi=80)
plt.title("Дискретный сигнал")
plt.stem(t2, wave, use_line_collection=True)
plt.xlim((-4, 4))
plt.grid(True)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 4), dpi=80)
plt.title("Квантованный сигнал")
plt.step(t2, wave)
plt.xlim((-4, 4))
plt.grid(True)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
plt.tight_layout()
```
